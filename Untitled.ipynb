#### Now we will deal with null values

```python
print('Columns having null values')
for i in df.columns:
    if df[i].isnull().sum()>0:
        print(f'{i} = {round((df[i].isnull().sum()/df.shape[0])*100,2)}')
```

**After removing the duplicate values still**
1. 'New', 'PendingS' have 100% null values so we will delete them.
2. 'fulfilled-by' only has 1 value and 70% of the value is null value.
3. 'currency','Amount' has 6.24% null values replace them
4. 'ship-city', 'ship-state', 'ship-postal-code' and 'ship-country' has 0.02% null values replace them.

```python
df.head(2)
```

```python
df.drop(columns=['New','PendingS',],inplace=True)
```

```python
df.drop(columns=['fulfilled-by'],inplace=True)
```

```python
print(df['currency'].value_counts())# We will replace all the null value of currency with INR only
print('-'*50)
print(df['Amount'].median())
```

```python
df['currency'].fillna(value='INR',inplace=True)
```

```python
sns.boxplot(x=df['Amount'])# It does have some outliers and here it will be useful
```

```python
amount_median = df['Amount'].median()
df['Amount'].fillna(amount_median,inplace=True)
```

```python
df['ship-city'].ffill(axis=0,inplace=True)
df['ship-state'].ffill(axis=0,inplace=True)
df['ship-postal-code'].ffill(axis=0,inplace=True)
df['ship-country'].ffill(axis=0,inplace=True)
```

```python
print('Columns having null values')
for i in df.columns:
    if df[i].isnull().sum()>0:
        print(f'{i} = {round((df[i].isnull().sum()/df.shape[0])*100,2)}')
```

```python
# We have removed all the null values sucessfully.
```
